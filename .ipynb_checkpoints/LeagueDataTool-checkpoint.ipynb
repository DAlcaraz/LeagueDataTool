```python
y = df['Armor']
```

```python
champFeatures = ['HP', 'HP per level', 'HP regen', 'Mana', 'Armor per level', 'MR', 'Range', 'HP regen', 'AD', 'AD per level', 'Attack speed']
```

```python
x = df[champFeatures]
```

```python
x.describe()
```

```python
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)
```

```python
champModel = DecisionTreeRegressor()
```

```python
champModel.fit(train_x, train_y)
```

```python
#Making predictions for Armor of
train_x
```

```python
predictions = champModel.predict(val_x)
print(predictions)
```

```python
val_y
```

```python
print(mean_absolute_error(val_y, predictions))
```
